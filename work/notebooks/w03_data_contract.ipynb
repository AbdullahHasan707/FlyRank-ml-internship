# 3. Feature Selection

```python
features = [
    "word_count",
    "content_age_days",
    "impressions_90d",
    "engagement_rate",
    "avg_position"
]

print("Selected Features:")

for feature in features:
    print(feature)
```

### Why these features?

| Feature | Reason |
|----------|--------|
| word_count | Represents the amount of content on a page. |
| content_age_days | Older pages are more likely to require updates. |
| impressions_90d | Measures search visibility. |
| engagement_rate | Indicates user interaction quality. |
| avg_position | Measures average Google ranking position. |

These features are observable before making a refresh decision and therefore do not leak future information.

# 4. Leakage Demonstration

print("Example of a leaky feature: trend_pct")

print("""
trend_pct is derived from future performance and therefore
would not be available when making a real prediction.

It must be excluded from model training.
""")

## Leakage Conclusion

Including **trend_pct** would artificially improve model performance because it contains information derived from the target.

Therefore it is removed from the feature set.
